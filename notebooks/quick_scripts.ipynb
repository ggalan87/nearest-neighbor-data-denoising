```python
# Mass rename
import os
from pathlib import Path
directory = '../lightning/embeddings/market1501'
orig_string = 'orig'
new_string = 'pop'

for p in sorted(Path(directory).iterdir(), key=os.path.getmtime):
    print(p)
    renamed = Path(str(p).replace(orig_string, new_string))
    p.rename(renamed)
    print(p)
    
```
